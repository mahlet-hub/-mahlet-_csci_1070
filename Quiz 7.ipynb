```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
credit_df = pd.read_csv('credit_card.csv')
credit_df.head()
```

```python
credit_df_encoded = pd.get_dummies(data = credit_df, dtype = int)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

credit_df_encoded.isnull().sum()
```

```python

X = np.array(credit_df_encoded['Employed_days']).reshape(-1,1)
y = np.array(credit_df_encoded['Type_Income_Pensioner']).reshape(-1, 1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 42)

# 'Employed_days','Type_Income_Pensioner'

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train) 
X_test = sc.fit_transform(X_test)

knn = KNeighborsClassifier(n_neighbors=13) 
knn.fit(X_train, y_train) 

y_pred = knn.predict(X_test)
print(y_pred)
print(type(y_pred))
```

```python
print(knn.score(X_test, y_test))
```

```python
import pandas as pd
from sklearn.feature_selection import VarianceThreshold


credit_df_encoded.fillna(credit_df_encoded.median(), inplace=True)


credit_df_encoded = pd.get_dummies(credit_df_encoded, drop_first=True)


threshold = 0.1
selector = VarianceThreshold(threshold=threshold)


credit_df_reduced = selector.fit_transform(credit_df_encoded)


selected_features = credit_df_encoded.columns[selector.get_support()]


print("Features after Variance :", selected_features)
print("Reduced Shape:", credit_df_reduced.shape)
```

```python
print(knn.score(X_test, y_test))
```
